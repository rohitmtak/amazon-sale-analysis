# Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Amazon Sale Report.csv", low_memory=False)
```

```python
# convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Status')
plt.title('Total Orders by Status')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Status', y='Amount', estimator=sum)
plt.title('Total Sales Amount by Status')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Fulfilment')
plt.title('Number of Orders by Fulfilment Method')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='ship-service-level')
plt.title('Number of Orders by Ship Service Level')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category')
plt.title('Number of Orders by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Category', y='Amount', estimator=sum)
plt.title('Total Sales Amount by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='ship-state', order=df['ship-state'].value_counts().index)
plt.title('Number of Orders by State')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.groupby('Date')['Amount'].sum().plot(kind='line')
plt.title('Total Sales Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Date'].value_counts().sort_index().plot(kind='line')
plt.title('Orders Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby(['Date', 'Fulfilment']).size().unstack().plot(kind='line')
plt.title('Number of Orders by Fulfilment Method over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby(['Date', 'Fulfilment'])['Amount'].sum().unstack().plot(kind='line')
plt.title('Total Sales Amount by Fulfilment Method over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby(['Date', 'Sales Channel ']).size().unstack().plot(kind='line')
plt.title('Number of Orders by Sales Channel over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby(['Date', 'Sales Channel '])['Amount'].sum().unstack().plot(kind='line')
plt.title('Total Sales Amount by Sales Channel over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby(['Date', 'ship-service-level']).size().unstack().plot(kind='line')
plt.title('Number of Orders by Ship Service Level over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby(['Date', 'Category']).size().unstack().plot(kind='line')
plt.title('Number of Orders by Category over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.show()
```
